```python
import pickle
%matplotlib inline
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import scipy
from scipy import stats
```

1.load the data and check for what are in the dataset

```python
Higgs = open ("higgs_100000_pt_1000_1200.pkl",'rb')
dict_Higgs = pickle.load(Higgs)
QCD = open ("qcd_100000_pt_1000_1200.pkl",'rb')
dict_QCD = pickle.load(QCD)
```

```python
dict_Higgs.keys()
```

The data is coming from ATLAS particle detector to find the standard model Higgs boson in pp collisions at âˆš s = 13 TeV at the LHC using simulated data. It has a signal file(higgs) and a background file(qcd).The data is measurement of jets that are emitted during proton-proton collisions. It has ['pt', 'eta', 'phi', 'mass', 'ee2', 'ee3', 'd2', 'angularity', 't1',
       't2', 't3', 't21', 't32', 'KtDeltaR']14 values that we can explore more.

2.Explore data

2.1 pt

```python
plt.hist(dict_Higgs['pt'],bins=100,alpha=.5,label='Higgs');
plt.hist(dict_QCD['pt'],bins=100,alpha=.5,label='QCD');
plt.title('pt')
plt.xlabel('pt')
plt.ylabel('Counts')
plt.legend();
```

pt is the transverse momentum of the jets,equation calculation is sqrt(px^2+py^2).Based on the histogramsthe higgs data showed a high spike at 1100, the qcd data doesn't have a clear spike, both has a rayleigh distribution.

2.2 eta

```python
plt.hist(dict_Higgs['eta'],bins=100,alpha=.5,label='Higgs');
plt.hist(dict_QCD['eta'],bins=100,alpha=.5,label='QCD');
plt.title('eta')
plt.xlabel('eta')
plt.ylabel('Counts')
plt.legend();
```

eta is the pseudorapidity of the jets,equation calculation is -ln(tan(theta/2)).Based on the histogramsthe higgs data showed smaller deviation than qcd data, both has a normal distribution

2.3 phi

```python
plt.hist(dict_Higgs['phi'],bins=100,alpha=.5,label='Higgs');
plt.hist(dict_QCD['phi'],bins=100,alpha=.5,label='QCD');
plt.title('phi')
plt.xlabel('phi')
plt.ylabel('Counts')
plt.legend();
```

phi is the Azimuthal Angle of the jets, the equation is cos^-1(x/r),both data has a very high noise and we can't tell by look at the histogram, which means this is not a correlate to find the result.

2.4 mass

```python
plt.hist(dict_Higgs['mass'],bins=100,alpha=.5,label='Higgs');
plt.hist(dict_QCD['mass'],bins=100,alpha=.5,label='QCD');
plt.title('mass')
plt.xlabel('mass')
plt.ylabel('Counts')
plt.xlim(0,400)
plt.legend();
```

Mass is the invariant mass of the jets, equation is E^2=p^2+m^2, E is the energy and p is the momentum. Higgs data has a high spike in M=130 and QCD showed a ralyleigh distribution spike at around 110.

2.5 ee2&ee3

```python
plt.subplots(1,2)
plt.subplot(1,2,1)
plt.hist(dict_Higgs['ee2'],bins=100,alpha=.5,label='Higgs');
plt.hist(dict_QCD['ee2'],bins=100,alpha=.5,label='QCD');
plt.title('ee2')
plt.xlabel('ee2')
plt.ylabel('ee2')
plt.legend()
plt.subplot(1,2,2)
plt.hist(dict_Higgs['ee3'],bins=200,alpha=.5,label='Higgs');
plt.hist(dict_QCD['ee3'],bins=200,alpha=.5,label='QCD');
plt.title('ee3')
plt.xlabel('ee3')
plt.ylabel('ee3')
plt.xlim(0,0.002)
plt.legend()
```

ee2 and ee3 represent the 2nd- 3rd energy-correlation functions Ratio. QCD has low ratio but lots of data, higgs has a higher ratio but smaller data.

2.6 d2

```python
plt.hist(dict_Higgs['d2'],bins=300,alpha=.5,label='Higgs');
plt.hist(dict_QCD['d2'],bins=300,alpha=.5,label='QCD');
plt.title('d2')
plt.xlabel('d2')
plt.ylabel('Counts')
plt.xlim(0,50)
plt.legend();
```

D2 is the ratio of ee2 and ee3 and it showed a very good graph of what we said about ee2 and ee3(QCD has low ratio but lots of data, higgs has a higher ratio but smaller data.)

2.7 angularity

```python
plt.hist(dict_Higgs['angularity'],bins=200,alpha=.5,label='Higgs');
plt.hist(dict_QCD['angularity'],bins=200,alpha=.5,label='QCD');
plt.title('angularity')
plt.xlabel('angularity')
plt.ylabel('Counts')
plt.xlim(0,0.015)

plt.legend();
```

The angularity measurement is a measurement of the radiation of the jet.For both data there is a spike at 0.001 and higgs has a larger counts than QCD.

2.8 t1&t2&t3

```python
plt.subplots(1,3)
plt.subplot(1,3,1)
plt.hist(dict_Higgs['t1'],bins=100,alpha=.5,label='Higgs');
plt.hist(dict_QCD['t1'],bins=100,alpha=.5,label='QCD');
plt.title('t1')
plt.xlabel('t1')
plt.ylabel('counts')
plt.legend()
plt.subplot(1,3,2)
plt.hist(dict_Higgs['t2'],bins=100,alpha=.5,label='Higgs');
plt.hist(dict_QCD['t2'],bins=100,alpha=.5,label='QCD');
plt.title('t2')
plt.xlabel('t2')
plt.ylabel('counts')
plt.legend()
plt.subplot(1,3,3)
plt.hist(dict_Higgs['t3'],bins=100,alpha=.5,label='Higgs');
plt.hist(dict_QCD['t3'],bins=100,alpha=.5,label='QCD');
plt.title('t3')
plt.xlabel('t3')
plt.ylabel('counts')
plt.legend()
```

t1 to t3 represent 1-2-3 Subjettiness of the jets, from t1 to t3 the higgs and qcd showed a more difference. Higgs always had a rayleigh distribution and qcd are approaching to normal distribution.

t21 is the ratio of t1 and t2, t32 is the ratio of t3 and t2, t21 had a more clear correcation and t32 doesn't.
